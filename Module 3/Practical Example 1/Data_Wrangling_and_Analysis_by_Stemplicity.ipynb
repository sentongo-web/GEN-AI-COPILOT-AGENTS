```python
# Pandas is a data manipulation and analysis library in Python that uses a data structure known as DataFrame
# A DataFrame is a two-dimensional tabular data structure with labeled rows and columns
# DataFrames are equivalent to Microsoft Excel spreadsheets
# Pandas is superior at automating data processing tasks than Excel and can work with a large amount of data
import pandas as pd
```

```python
# Read the Excel file containing client demographics
df_demographics = pd.read_excel('Client_Demographics.xlsx')
df_demographics
```

```python
# Read the Excel file containing client Financial Data
df_financials = pd.read_excel('Client_Financial_Details.xlsx')
df_financials
```

# Task 2. Handle Missing Values

image.png

```python
# Use the .isnull() method to find missing values in the Pandas DataFrame
df_demographics.isnull()
```

```python
# Add up all missing values
df_demographics.isnull().sum()
```

image.png

```python
# Fill out the missing values in the "Age" column with the average age
df_demographics['Age'] = df_demographics['Age'].fillna(df_demographics['Age'].mean())
df_demographics
```

```python
# Fill missing values in categorical columns with the mode
# Note that New York is the most frequent city
df_demographics['City'] = df_demographics['City'].fillna(df_demographics['City'].mode()[0])
df_demographics
```

```python
# Let's check if we still have any missing values
df_demographics.isnull().sum()
```

```python
# Use the .isnull() method to find missing values in the df_financials Pandas DataFrame
df_financials.isnull()
```

```python
# Fill in missing values in the df_financials Pandas DataFrame
df_financials['Risk_Tolerance'] = df_financials['Risk_Tolerance'].fillna(df_financials['Risk_Tolerance'].mode()[0])
df_financials
```

# Task 3. Merge Pandas DataFrames

image.png

```python
# Merge the two Pandas DataFrames "df_demographics" and "df_financials"
# Note that left join returns all rows from the left DataFrame, and any rows with matching keys from the right DataFrame.
# When there is no matching from right DataFrame, NaN will be returned
df_clients_combined = pd.merge(df_demographics, df_financials, on = 'Client_ID', how = 'left')
df_clients_combined
```

# Task 4. Perform Data Analysis, Filtering, & Sorting

```python
# Obtain a statistical summary
df_clients_combined.describe()
```

```python
# Filter out the Pandas DataFrame to obtain high net worth clients with over 150,000
high_networth_clients = df_clients_combined[df_clients_combined['Investment_Portfolio_Size'] > 150000]
high_networth_clients
```

```python
# Sort the Pandas DataFrame in an ascending order based on the "Investment_Portfolio_Size" column
df_clients_combined.sort_values(by = 'Investment_Portfolio_Size')
```

# Task 5. Visualize Datasets

image.png

- Link to Matplotlib Documentation: https://matplotlib.org/
- Link to Seaborn Documentation: https://seaborn.pydata.org/
- Link to Plotly Documentation: https://plotly.com/python/

```python
# Let's import data visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# PLot a Scatter plot showing Age vs. Investment Portfolio Size
plt.figure(figsize=(8, 5))
plt.scatter(df_clients_combined["Age"], df_clients_combined["Investment_Portfolio_Size"], alpha=0.75, s=100)
plt.title("Age vs. Investment Portfolio Size")
plt.xlabel("Age (Years)")
plt.ylabel("Investment Portfolio Size ($)")
plt.grid(True)
```

```python
# Plot the histogram of the Age column using 3 bins
plt.figure(figsize=(7, 5))
df_clients_combined['Age'].plot(kind='hist', bins = 3)
plt.title('Age Histogram')
plt.xlabel('Age')
```

```python
# Group data by advisor and sum up Portfolio_Value
advisor_portfolio = df_clients_combined.groupby('Advisor')['Investment_Portfolio_Size'].sum()

# Create a bar chart
plt.figure(figsize=(7, 5))
advisor_portfolio.plot(kind='bar')
plt.title('Total Portfolio Value by Advisor')
plt.xlabel('Advisor')
plt.ylabel('Total Portfolio Value')
plt.xticks(rotation=45)
```

```python
# Heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(df_clients_combined.corr(numeric_only=True), annot=True)
plt.title("Heatmap of Correlation Matrix")
plt.show()
```
